```python
import pandas as pd
import numpy as np

df = pd.read_csv("abel-est-region-flow-2005.xlsx - flow estimates by region 2005.csv")
df
```

```python
df = df.iloc[: , 1:]
df.columns = df.iloc[0]
df = df.iloc[1: , :]
df
```

```python
df = df.set_index(np.nan)
```

```python
df.to_csv("altered.csv")
```

```python
new = pd.read_csv("altered.csv")
new
```

```python
df_matrix = new.stack().reset_index().rename(columns={'level_0':'Source','level_1':'Target', 0:'Weight'})
df_matrix
```

```python
df_matrix = df_matrix.loc[df_matrix['Target'] != "nan"]
df_matrix.head(20)
```

```python
df_matrix['Target'] = df_matrix['Target'].map({
                         'North America': 0, 
                         'Central America': 1, 
                         'South America': 2, 
                         'North Africa': 3, 
                         'Sub-Saharan Africa': 4, 
                         'Northern Europe': 5, 
                         'Western Europe': 6, 
                         'Southern Europe': 7, 
                         'Eastern Europe': 8, 
                         'Central Asia': 9, 
                         'Western Asia': 10,
                         'South Asia': 11,
                         'East Asia': 12,
                         'South-East Asia': 13,
                         'Oceania': 14
                        })
```

```python
df_matrix['Weight'] = df_matrix['Weight'].apply(lambda x: x.replace(',', ''))
```

```python
df_matrix['Weight'] = pd.to_numeric(df_matrix['Weight'])
```
