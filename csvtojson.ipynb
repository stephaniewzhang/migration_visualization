```python
import pandas as pd
import numpy as np

df = pd.read_csv("abel-est-region-flow-2005.xlsx - flow estimates by region 2005.csv")
df
```

```python
df_matrix = df.stack().reset_index().rename(columns={'level_0':'Source','level_1':'Target', 0:'Weight'})
df_matrix
```

```python
print(df_matrix.shape)
df_matrix = df_matrix.loc[df_matrix['Target'] != "Sub-Continent"]
print(df_matrix.shape)
df_matrix
```

```python
df_matrix['Target'] = df_matrix['Target'].map({
    'North America_United States':0,
    'North America_Canada':1,
    'Latin America': 2,
    'Africa': 3,
    'Europe':4,
    'Fmr Soviet Union': 5, 
    'West Asia': 6,
    'South Asia': 7,
    'East Asia_Hong Kong and Macao': 8,
    'East Asia_China': 9,
    'East Asia_Japan':10,
    'East Asia_North Korea':11,
    'East Asia_South Korea':12,
    'East Asia_Mongolia':13,
    'South-East Asia':14,
    'Oceania':15})
```

```python
df_matrix
```

```python
df_matrix['Weight'] = pd.to_numeric(df_matrix['Weight'])
```

```python
#calculate the weight of each of the nodes
nodes = []

for num in range(16):
    subcontinent = ['United States','Canada','Latin America','Africa','Europe',
                    'Fmr Soviet Union','West Asia','South Asia','Hong Kong and Macao','China',
                    'Japan','North Korea','South Korea','Mongolia','South-East Asia','Oceania']
    
    sum_source = df_matrix.loc[df_matrix['Source'] == num]['Weight'].sum()
    sum_target = df_matrix.loc[df_matrix['Target'] == num]['Weight'].sum()
    sum_total = sum_source + sum_target
    
    if (num < 2):
        continent = "North America"
    elif (num < 3):
        continent = "Latin America"
    elif (num < 4):
        continent = "Africa"
    elif (num < 5):
        continent = "Europe"
    elif (num < 6):
        continent = "Fmr Soviet Union"
    elif (num < 7):
        continent = "West Asia"
    elif (num < 8):
        continent = "South Asia"
    elif (num < 14):
        continent = "East Asia"
    elif (num < 15):
        continent = "South-East Asia"
    else:
        continent = "Oceania"
    node = [continent, subcontinent[num], sum_total]
    nodes.append(node)

nodes
```

```python
df_nodes = pd.DataFrame(nodes, columns=['SubContinent', 'Country', 'Weight'])
```
