```python
def lifeScale (n):
    return ((n - lifeMin) * 10) / (lifeMax-lifeMin)
```

```python
def popScale (n):
    return ((n - popMin) * 10) / (popMax-popMin)
```

```python
def gdpScale(n):
     return ((n - gdpMin) * 10) / (gdpMax-gdpMin)
```

```python
def genderGapScale(n):
     return ((n - genderGapMin) * 10) / (genderGapMax-genderGapMin)
```
